```python
import pandas as pd 
input_path="Policy Data.xlsx"
input_data = pd.read_excel(input_path)
print(input_data)
```

```python
DESIRED_COLUMNS = [
    'Full Name', 'Date of Birth', 'Age', 'Gender', 'Marital Status', 'Contact Information',
    'Employer Name', 'Employee Designation', 'Employment Status', 'Annual Income (₹)',
    'Joining Date with Employer', 'Policy Number', 'Premium Payment Frequency', 'Policy Type',
    'Coverage Amount (₹)', 'Spouse Covered', 'Parents Covered', 'Medical History', 'Current Health Status'
]
print(DESIRED_COLUMNS)
```

```python
prompt = f"""
I am sending you raw, incomplete, or inconsistent tabular data. Your task is to clean and organize the data, handle missing or placeholder values, calculate any missing fields (like "Age"), and ensure everything is consistent. Output the cleaned data as a JSON array with the correct values. 

Input Data: {input_data}

Return only the JSON result without any additional explanation or code.

Desired columns ={DESIRED_COLUMNS}
"""
```

```python
import openai
groq_api_key="groq_api_key"
client = openai.OpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=groq_api_key
)
response = client.chat.completions.create(
    model="llama3-70b-8192",  # Or another Groq-supported model
    messages=[
        {"role": "user", "content": prompt}
    ]
)

print(response.choices[0].message.content)
```
